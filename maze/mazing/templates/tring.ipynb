```python
from music21 import converter, instrument

# Load the MusicXML file
file_path = "C://Users//Asamoah//Downloads//10.xml"
score = converter.parse(file_path)

# Dictionary to hold notes and durations for each part
parts_data = {
    'Soprano': [],
    'Alto': [],
    'Tenor': [],
    'Bass': []
}

# Helper function to get the part name
def get_part_name(part):
    part_name = None
    for p in part.getElementsByClass(instrument.Instrument):
        part_name = p.partName
    return part_name

# Extract notes and durations for each part
for part in score.parts:
    part_name = get_part_name(part)
    if part_name in parts_data:
        for note in part.flat.notes:
            note_name = note.nameWithOctave
            duration = note.duration.quarterLength
            parts_data[part_name].append((note_name, duration))

# Print the extracted notes and durations
for part, notes in parts_data.items():
    print(f"\n{part} Notes and Durations:")
    for note, duration in notes:
        print(f"Note: {note}, Duration: {duration}")
```

```python
import pandas as pd

notes = [
    ['F4 1.0', 'F4 1.0', 'F3 1.0', 'F3 1.0'],
    ['B-4 1.0', 'B-4 1.0', 'B-3 1.0', 'B-3 1.0'],
    ['F4 1.0', 'F4 1.0', 'B-3 1.0', 'D3 1.0'],
    ['B-4 1.0', 'G4 0.5', 'E-3 0.5', 'E-3 0.5'],
    ['-', 'F4 0.5', 'F3 0.5', 'D3 0.5'],
    ['C5 1.0', 'E-4 1.0', 'G3 0.5', 'C3 1.0'],
    ['A3 0.5'],
    ['D5 1.5', 'D4 1.5', 'B-3 1.5', 'B-2 1.5'],
    ['C5 0.5', 'E-4 0.5', 'F3 0.5', 'B-2 0.5'],
    ['B-4 1.0', 'D4 1.0', 'F3 1.0', 'B-2 1.0']
]

# Create DataFrame
df = pd.DataFrame(notes, columns=['Note1', 'Note2', 'Note3', 'Note4'])

# Function to extract duration from note string
def extract_duration(note):
    if isinstance(note, str):
        parts = note.split()
        if len(parts) == 2 and parts[1].replace('.', '').isdigit():
            return float(parts[1])
    return 0.0

# Add cumulative duration columns
for col in df.columns:
    df[col + '_Cumulative'] = df[col].apply(extract_duration).cumsum()

# Display the updated DataFrame
print(df)
```

```python
# Function to check for the index of the first sublist with different duration values
def find_first_different_duration_index(notes):
    for index, sublist in enumerate(notes):
        durations = [extract_duration(note) for note in sublist]
        if len(set(durations)) > 1:
            return index
    return -1

# Find the index
index = find_first_different_duration_index(notes)
print(f"The index of the first sublist with different duration values is: {index}")
```

```python
from music21 import note, stream

# Given music notes
music_notes = [
    ['E-4 1.0', 'B-3 1.0', 'G3 1.0', 'E-3 1.0'],
    ['E-4 0.5', 'B-3 0.5', 'G3 0.5', 'E-3 0.5'],
    ['E-4 0.5', 'B-3 0.5', 'G3 0.5', 'E-3 0.5'],
    ['G4 1.0', 'E-4 1.0', 'G3 1.0', 'C3 1.0'],
    ['G4 1.0', 'E-4 1.0', 'G3 1.0', 'C3 1.0'],
    ['C5 3.0', 'E-4 3.0', 'A-3 3.0', 'A-2 3.0'],
    ['C5 1.0', 'E-4 1.0', 'A-3 1.0', 'A-3 1.0']
]

# Create a stream
melody = stream.Stream()

# Extract and add the first note from each sublist
for sublist in music_notes:
    note_str = sublist[3]  # First member of each sublist
    pitch, duration = note_str.split()  # Split into pitch and duration
    duration = float(duration)  # Convert duration to float
    # Create a note object and add it to the stream
    n = note.Note(pitch)
    n.quarterLength = duration
    melody.append(n)

# Show (and play) the stream
melody.show('midi')
```

```python
from music21 import converter, stream, midi, note, chord

# Load the two XML files
score1 = converter.parse(r'C:\Users\KOFI ADUKPO\Desktop\code\aseda\adom\templates\tunes_ang\1.xml')
score2 = converter.parse(r'C:\Users\KOFI ADUKPO\Desktop\code\aseda\adom\templates\tunes_ang\2.xml')

# Create a new Stream to hold the combined scores
combined_stream = stream.Stream()

# Append notes and rests from the first score
for elem in score1.flat.notesAndRests:
    combined_stream.append(elem)

# Offset the second score to start after the first score ends
end_time = score1.highestTime
for elem in score2.flat.notesAndRests:
    if isinstance(elem, note.Note) or isinstance(elem, note.Rest):
        # Create a new note or rest with the adjusted offset
        new_elem = elem.__class__(name=elem.name, quarterLength=elem.quarterLength, pitch=elem.pitch if hasattr(elem, 'pitch') else None)
        new_elem.offset = elem.offset + end_time
    elif isinstance(elem, chord.Chord):
        # Create a new chord with the adjusted offset
        new_elem = chord.Chord(pitches=elem.pitches, quarterLength=elem.quarterLength)
        new_elem.offset = elem.offset + end_time
    else:
        # Handle other elements if necessary
        continue
    combined_stream.append(new_elem)

# Sort the combined stream
combined_stream.sort()

# Save the combined stream as a MIDI file
combined_midi_path = r'C:\Users\KOFI ADUKPO\Desktop\code\aseda\adom\templates\tunes_ang\combined.mid'
mf = midi.translate.music21ObjectToMidiFile(combined_stream)
mf.open(combined_midi_path, 'wb')
mf.write()

print(f"Combined MIDI file saved to: {combined_midi_path}")

# To play the combined score, you can use a MIDI player or open the file in a media player that supports MIDI
```
