```python
import os
import re

# Specify the directory containing the HTML files
directory = r'C:\Users\KOFI ADUKPO\Downloads\lyrics'

# Define the HTML snippet to be added
snippet = '<a href="/">Read and play tonic-solfa notes</a>'

# Function to process an HTML file
def process_html_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    # Add the snippet before the closing </body> tag
    modified_content = re.sub(r'</body>', f'{snippet}\n</body>', content, flags=re.IGNORECASE)

    with open(file_path, 'w') as file:
        file.write(modified_content)

# Iterate through HTML files in the directory
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith('.html'):
            file_path = os.path.join(root, file)
            process_html_file(file_path)
            print(f'Modified: {file_path}')

print('All HTML files processed.')
```

```python
import os
from bs4 import BeautifulSoup

# Directory path
directory = r"C:\Users\KOFI ADUKPO\Downloads\lyrics"

# Iterate through files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".html"):
        file_path = os.path.join(directory, filename)

        # List of encoding types to try
        encodings = ['utf-8', 'latin-1', 'windows-1252']  # Add more as needed

        content = None
        # Attempt to read the file with different encodings
        for encoding in encodings:
            try:
                with open(file_path, 'r', encoding=encoding) as file:
                    content = file.read()
                    break  # Break the loop if the file is successfully read
            except UnicodeDecodeError:
                pass  # If decoding fails, try the next encoding

        if content is not None:
            soup = BeautifulSoup(content, 'html.parser')

            # Retrieve all items in the h1 tag and store them in 'kog'
            h1_tags = soup.find_all('h1')
            kog = [h1.get_text() for h1 in h1_tags]
            # Replace "\n" with "<br>"
            kog = kog[0].replace("\n", " <br> ")

            # print(p_with_br)

            # Replace content of h1 tags with the 'kog' variable
            for h1 in h1_tags:
                h1.string = ' '.join(kog)

            # Update the file with the modified content
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(str(soup))

print("Replacement completed in all HTML files.")
```

```python
def rearrange_list(input_list):
    def extract_integer(s):
        # Extracts the integer part from the string
        num = ''
        for char in s:
            if char.isdigit():
                num += char
            else:
                break
        return int(num) if num else None

    # Custom sorting function based on extracted integers
    def sort_key(elem):
        return extract_integer(elem)

    # Sort the list based on extracted integers
    sorted_list = sorted(input_list, key=sort_key)
    return sorted_list

# Example list
original_list = file_list

# Rearrange the list
rearranged_list = rearrange_list(original_list)
print(rearranged_list)
```

```python
#removing those special characters

import os
import codecs

directory_path = r"C:\Users\KOFI ADUKPO\Downloads\lyrics"

def remove_non_utf8_characters(file_path):
    try:
        with codecs.open(file_path, 'r', encoding='ISO-8859-1') as file:
            content = file.read()
        # Reopen the file in write mode with utf-8 encoding to write the content back
        with codecs.open(file_path, 'w', encoding='utf-8') as file:
            file.write(content)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)
    # Check if the file is an HTML file
    if os.path.isfile(file_path) and file_path.lower().endswith('.html'):
        remove_non_utf8_characters(file_path)
```

```python
import os

def clean_html_file(input_file, output_file, encoding='utf-8'):
    try:
        with open(input_file, 'rb') as f:
            # Read the binary content of the file
            content = f.read()

        # Decode the content using the specified encoding
        decoded_content = content.decode(encoding, errors='replace')

        # Encode the content back to bytes using the same encoding
        encoded_content = decoded_content.encode(encoding, errors='replace')

        # Write the cleaned content to the output file
        with open(output_file, 'wb') as f:
            f.write(encoded_content)

        print(f"File cleaned and saved as {output_file}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

def clean_html_files_in_directory(directory_path, output_directory, encoding='utf-8'):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Iterate through all files in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith('.html'):
            input_file_path = os.path.join(directory_path, filename)
            output_file_path = os.path.join(output_directory, filename)
            clean_html_file(input_file_path, output_file_path, encoding)

# Replace the paths with your actual directory paths
input_directory = r'C:\\Users\\KOFI ADUKPO\\Desktop\\code\\aseda\\adom\\templates\\lyrics'
output_directory = r'C:\\Users\\KOFI ADUKPO\\Desktop\\code\\aseda\\adom\\templates\\lyrics'

clean_html_files_in_directory(input_directory, output_directory)
```
