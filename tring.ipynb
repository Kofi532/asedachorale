```python
import pandas as pd

notes = [
    ['F4 1.0', 'F4 1.0', 'F3 1.0', 'F3 1.0'],
    ['B-4 1.0', 'B-4 1.0', 'B-3 1.0', 'B-3 1.0'],
    ['F4 1.0', 'F4 1.0', 'B-3 1.0', 'D3 1.0'],
    ['B-4 1.0', 'G4 0.5', 'E-3 0.5', 'E-3 0.5'],
    ['F4 0.5', 'F3 0.5', 'D3 0.5'],
    ['C5 1.0', 'E-4 1.0', 'G3 0.5', 'C3 1.0'],
    ['A3 0.5'],
    ['D5 1.5', 'D4 1.5', 'B-3 1.5', 'B-2 1.5'],
    ['C5 0.5', 'E-4 0.5', 'F3 0.5', 'B-2 0.5'],
    ['B-4 1.0', 'D4 1.0', 'F3 1.0', 'B-2 1.0']
]

# Create DataFrame
df = pd.DataFrame(notes, columns=['Note1', 'Note2', 'Note3', 'Note4'])

# Function to extract duration from note string
def extract_duration(note):
    if isinstance(note, str):
        parts = note.split()
        if len(parts) == 2 and parts[1].replace('.', '').isdigit():
            return float(parts[1])
    return 0.0

# Add cumulative duration columns
for col in df.columns:
    df[col + '_Cumulative'] = df[col].apply(extract_duration).cumsum()

# Display the updated DataFrame
print(df)
```

```python
from music21 import stream, note, chord

# Create a stream to hold the notes
score = stream.Stream()

# Define the notes and durations for the first part
notes = ['F4 1.0', 'F4 1.0', 'F#3 1.0']

# Define the harmony note and its duration
harmony_note = 'C#5'
harmony_duration = 5.0

# Create a list to hold the combined notes
combined_notes = []

# Add the first part notes to the combined list
for n in notes:
    pitch, duration = n.split()
    combined_notes.append(note.Note(pitch, quarterLength=float(duration)))

# Add the harmony note to the combined list
combined_notes.append(note.Note(harmony_note, quarterLength=harmony_duration))

# Create chords from the combined notes for harmony
for i in range(len(notes)):
    # Create a chord with the current note and the harmony note
    chord_notes = [combined_notes[i], combined_notes[-1]]
    new_chord = chord.Chord(chord_notes)
    score.append(new_chord)

# Show the score and play the notes
score.show('midi')
```

```python
from music21 import stream, note, midi

# Create a stream
s = stream.Stream()

# Define the solfa notes
solfa_notes = ['d', 'r', 'm', 'f', 's', 'l', 't', 'd']

# Map solfa notes to MIDI pitches (assuming C major for simplicity)
solfa_to_pitch = {
    'd': 60,  # C
    'r': 62,  # D
    'm': 64,  # E
    'f': 65,  # F
    's': 67,  # G
    'l': 69,  # A
    't': 71,  # B
    'd>': 72  # C (octave higher)
}

# Add notes to the stream
for solfa in solfa_notes:
    s.append(note.Note(solfa_to_pitch[solfa], quarterLength=1))

# Write the stream to a MIDI file
mf = midi.translate.music21ObjectToMidiFile(s)
output_path = 'C:\\Users\\Asamoah\\Desktop\\kofi\\maze\\mazing\\static\\solfa_notes.mid'
mf.open(output_path, 'wb')
mf.write()
mf.close()
```

```python
import mido

def is_valid_midi(file_path):
    try:
        with open(file_path, 'rb') as f:
            header = f.read(4)
            return header == b'MThd'
    except Exception as e:
        print(f"Error checking file {file_path}: {e}")
        return False

def merge_midi_files(file1, file2, output_file):
    # Check if both files are valid MIDI files
    if not is_valid_midi(file1):
        raise OSError(f"{file1} is not a valid MIDI file or cannot be read.")
    if not is_valid_midi(file2):
        raise OSError(f"{file2} is not a valid MIDI file or cannot be read.")
    
    # Load the MIDI files
    try:
        mid1 = mido.MidiFile(file1)
        mid2 = mido.MidiFile(file2)
    except Exception as e:
        raise OSError(f"Failed to read MIDI files: {e}")
    
    # Create a new MIDI file for the output
    merged_midi = mido.MidiFile()
    merged_track = mido.MidiTrack()
    merged_midi.tracks.append(merged_track)
    
    # Add all messages from the first file
    for track in mid1.tracks:
        for msg in track:
            merged_track.append(msg)
    
    # Add all messages from the second file
    for track in mid2.tracks:
        for msg in track:
            merged_track.append(msg)
    
    # Save the merged MIDI file
    try:
        merged_midi.save(output_file)
        print(f"Merged MIDI file saved as {output_file}")
    except Exception as e:
        raise OSError(f"Failed to save merged MIDI file: {e}")

# Define the paths to the MIDI files
file1 = r"C:\Users\Asamoah\Desktop\kofi\aseda\media\inn\output_file1.wav"
file2 = r"C:\Users\Asamoah\Desktop\kofi\aseda\media\inn\output_file2.wav"
output_file = r"C:\Users\Asamoah\Desktop\kofi\aseda\media\inn\merged_recording.mid"

# Merge the MIDI files
merge_midi_files(file1, file2, output_file)
```
